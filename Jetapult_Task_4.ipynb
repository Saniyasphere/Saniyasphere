```
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```
def detect_outliers_zscore(df, threshold=3):
    z_scores = (df - df.mean()) / df.std()
    return np.abs(z_scores) > threshold

numerical_columns = df.select_dtypes(include=[np.number]).columns
for col in numerical_columns:
    outliers = detect_outliers_zscore(df[col])
    print(f"Outliers in {col}: {outliers.sum()}")
```

```
plt.figure(figsize=(10, 6))
for i, col in enumerate(numerical_columns):
    plt.subplot(len(numerical_columns), 1, i + 1)
    sns.boxplot(x=df[col], color='lightblue')
    plt.title(f"Boxplot of {col}")
plt.tight_layout()
plt.show()
```

```
plt.figure(figsize=(6, 4))
sns.scatterplot(x=df['level_time'], y=df['session_duration'], alpha=0.6, color='blue')
plt.title("Scatter Plot: level_time vs session_duration")
plt.xlabel("Level Time")
plt.ylabel("Session Duration")
plt.show()
```

```
plt.figure(figsize=(8, 5))
sns.boxplot(x=df['event_name'], y=df['session_duration'], palette='Set3')
plt.title("Boxplot: session_duration by event_name")
plt.xticks(rotation=45)
plt.xlabel("Event Name")
plt.ylabel("Session Duration")
plt.show()
```

```
# Select numerical columns
numerical_columns = ['level_number', 'lvlcompstatus', 'level_time',
                     'source_amount', 'sink_amount',
                     'session_duration', 'total_duration_per_user']

plt.figure(figsize=(10, 8))
correlation_matrix = df[numerical_columns].corr()
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f", linewidths=0.5)
plt.title("Correlation Heatmap")
plt.show()
```

```
plt.figure(figsize=(6, 4))
sns.scatterplot(x=df['source_amount'], y=df['sink_amount'], alpha=0.6, color='green')
plt.title("Scatter Plot: source_amount vs sink_amount")
plt.xlabel("Source Amount")
plt.ylabel("Sink Amount")
plt.show()
```

```
plt.figure(figsize=(8, 5))
sns.boxplot(x=df['platform'], y=df['session_duration'], palette='coolwarm')
plt.title("Boxplot: session_duration by platform")
plt.xticks(rotation=45)
plt.xlabel("Platform")
plt.ylabel("Session Duration")
plt.show()
```

```
plt.figure(figsize=(8, 6))
df['session_duration'].hist(bins=30, color='skyblue', edgecolor='black')
plt.title("Histogram: Session Duration")
plt.xlabel("Session Duration")
plt.ylabel("Frequency")
plt.show()
```

```
# Grouping data by event_date and summing session_duration
daily_sessions = df.groupby('event_date')['session_duration'].sum()

plt.figure(figsize=(12, 6))
daily_sessions.plot(kind='line', color='blue', linewidth=2)
plt.title("Line Graph: Total Session Duration Over Time")
plt.xlabel("Event Date")
plt.ylabel("Total Session Duration")
plt.grid(True)
plt.show()
```

```
# Grouping data by event_date and calculating average level_time
daily_level_time = df.groupby('event_date')['level_time'].mean()

plt.figure(figsize=(12, 6))
daily_level_time.plot(kind='line', color='green', linewidth=2)
plt.title("Line Graph: Average Level Time Over Time")
plt.xlabel("Event Date")
plt.ylabel("Average Level Time")
plt.grid(True)
plt.show()
```
